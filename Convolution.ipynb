# Cross-Correlation (aka "Convolution")

```python
import skimage.io
import numpy as np
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
from scipy.signal import fftconvolve
```

First, let's load in an image of Layla Ophelia Litovsky (LOL)

```python
I = skimage.io.imread("layla.png")
I = rgb2gray(I)
I = I[200:400, 400:600]
plt.figure(figsize=(6, 6))
plt.imshow(I, cmap='gray')
```

Now let's create a "convolution kernel"

```python
kernel = np.ones((3, 3))

kernel = np.array([[-1, 0.0001, 1],
                    [-1, 0.0001, 1],
                    [-1, 0.0001, 1]]).T

kernel /= np.sum(kernel)
print(kernel)
plt.figure()
plt.imshow(kernel, cmap='magma')


res = fftconvolve(I, np.fliplr(np.flipud(kernel)), 'same')
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.imshow(I, cmap='gray')
plt.title("Before")
plt.subplot(122)
plt.imshow(res, cmap='gray')
plt.title("After")
```
